```python
import numpy
import random
#import libraries
X = [2,4,6,8,10]
#Training Data
Y = [4,8,12,16,20]
#Training Lables
w = random.random()
b = random.random()
#Initializing Weights and Bias Randomly
learningrate = 0.01

epochs = 1000
def error(y,prediction):
    return (y-prediction)**2
#Mean Squared Error Loss
def forwardPropagation(x,weight,bias):
    prediction = weight*x+bias
    return prediction

def backPropagation(y,prediction,x,weight,bias,learningrate):
    loss = error(y,prediction)
    weight-=-1*(2*(y-prediction)*x)*learningrate
    
    bias-=-1*(2*(y-prediction))*learningrate
    #Taking the derivative of the loss with respect weight and bias
    return loss,weight,bias

    
    
def train(X,Y,w,b,learningrate,epochs):
    
    for i in range(epochs):
        print("Epoch : "+str(i))
        Error = 0
        for j in range(len(X)):
            prediction = forwardPropagation(X[j],w,b)
            loss,w,b = backPropagation(Y[j],prediction,X[j],w,b,learningrate)
            print("Batch No"+str(j)+" Error : "+str(loss))
            Error+=loss
        print("Average Error : "+str(Error/len(X)))
    print(w,b)
train(X,Y,w,b,learningrate,epochs)
        
            
    
    
    
```
